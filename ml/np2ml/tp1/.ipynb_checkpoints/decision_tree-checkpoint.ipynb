# Decision Tree - ID3 algo

```python
import math
import itertools

import numpy as np
import pandas as pd
import ete3
```

```python
df = pd.read_csv('weather.csv')
df.head(n=len(df))
```

```python
class Tree:
    def __init__(self, attr, val, root=False, ete3_node=None):
        self.attr = attr
        self.val = val
        self.root = root
        if root:
            self.ete3_node = ete3.Tree(name='Root')
        else:
            self.ete3_node = ete3_node
        
        self.children = []
        
        
    def add_child(self, attr, val, df=None, target=None):
        if attr is None:
            distribution = round(len(df[df[target] == val]) / len(df), 2) * 100
            msg = 'Leaf: {}% {} '.format(int(distribution), val)
        else:
            msg = '{} == {}'.format(attr, val)
            
        if len(self.children) == 0:
            ete3_child =  self.ete3_node.add_child(name=msg)
        else:
            ete3_child = self.children[-1].ete3_node.add_sister(name=msg)
            ete3_child.add_face(ete3.TextFace(msg), column=0, position = "branch-right")
        
        node = Tree(attr, val, ete3_node=ete3_child)
        self.children.append(node)
        
        return node
    

class ID3:
    def __init__(self, criterion='entropy'):
        self.criterion = criterion
        
        
    def fit(self, df, target_attr, attrs, criterion='entropy'):
        self.df = df
        self.root = Tree(None, None, root=True)
        self.build_tree(self.root, df, target_attr, attrs)
        
    
    def build_tree(self, node, df, target_attr, attrs):
        if len(df) == 0:
            raise ValueError('Something went wrong')
        elif len(attrs) == 0 or df[target_attr].value_counts().max() == len(df): # Leaf
            node.add_child(None, df[target_attr].value_counts().idxmax(), df=df, target=target_attr)
            return
        
        if self.criterion == 'entropy':
            best_attr_idx = self.get_min_entropy(df, attrs)
        elif self.criterion == 'infogain':
            best_attr_idx = self.get_max_info_gain(df, target_attr, attrs)
        elif self.criterion == 'gainratio':
            best_attr_idx = self.get_max_gain_ratio(df, target_attr, attrs)
        else:
            raise ValueError('Unknown criterion {}.'.format(self.criterion))
            
        best_attr = attrs[best_attr_idx]
        
        remaining_attrs = attrs[:best_attr_idx] + attrs[best_attr_idx+1:]
        
        for val in df[best_attr].unique():
            filtered_examples = df[df[best_attr] == val]
            child = node.add_child(best_attr, val)
            self.build_tree(child, filtered_examples, target_attr, remaining_attrs)
    
    
    def predict(self, df_example):
        node = self.root
        while True:
            if len(node.children) == 1 and len(node.children[0].children) == 0:
                return node.children[0].val
            
            for child in node.children:
                if df_example[child.attr] == child.val:
                    node = child
                    break
            
    
    def display_tree(self):
        ts = ete3.TreeStyle()
        ts.show_scale = False
        ts.show_leaf_name = True
        return self.root.ete3_node.render('%%inline', w=200, units='mm', tree_style=ts)
    
    
    def score(self, df, target):
        nb_elts = len(df)
        good_preds = 0
        for i in range(nb_elts):
            pred, true = self.predict(df.iloc[i]), df.iloc[i][target]
            if pred == true:
                good_preds += 1
                
        return good_preds / nb_elts
    
    
    def get_min_entropy(self, df, attrs):
        entropies = np.zeros(len(attrs))
        nb_elts = len(df)
        for i, attr in enumerate(attrs):
            entropies[i] = self.compute_entropy(df, nb_elts, attr)
            
        return entropies.argmin()


    def compute_entropy(self, df, nb_elts, attr):
        unique_vals = df[attr].unique()
        entropy = 0
        for x in unique_vals:
            p = len(df[df[attr] == x]) / nb_elts
            entropy += -p * math.log2(p)
        return entropy
    

    def get_max_info_gain(self, df, target_attr, attrs):
        info_gains = np.zeros(len(attrs))
        nb_elts = len(df)
        for i, attr in enumerate(attrs):
            info_gains[i] = self.compute_info_gain(df, nb_elts, target_attr, attr)
        return info_gains.argmax()
    
    
    def compute_info_gain(self, df, nb_elts, target_attr, attr):
        attr_vals = list(df[attr].unique())
        target_vals = list(df[target_attr].unique())
        info_gain = 0
        
        for (attr_val, target_val) in itertools.product(attr_vals, target_vals):
            pac = len(df[(df[attr] == attr_val) & (df[target_attr] == target_val)]) + 1e-8
            pa = len(df[df[attr] == attr_val])
            pc = len(df[df[target_attr] == target_val])
            
            info_gain += pac * math.log2(pac / (pa * pc))
            
        return info_gain
    
    
    def get_max_gain_ratio(self, df, target_attr, attrs):
        gain_ratios = np.zeros(len(attrs))
        nb_elts = len(df)
        for i, attr in enumerate(attrs):
            gain_ratios[i] = self.compute_gain_ratio(df, nb_elts, target_attr, attr)
        return gain_ratios.argmax()
    
    
    def compute_gain_ratio(self, df, nb_elts, target_attr, attr):
        info_gain = self.compute_info_gain(df, nb_elts, target_attr, attr)
        split_info = 0
    
        for attr_val in list(df[attr].unique()):
            si = len(df[df[attr] == attr_val])
            ratio = si / nb_elts
            split_info += ratio * math.log2(ratio)
            
        return -split_info
```

```python
target_attr = df.columns.tolist()[-1]
attrs = df.columns.tolist()[:-1]

id3 = ID3(criterion='infogain')
id3.fit(df, target_attr, attrs)
id3.score(df, target_attr)
```

```python
id3.display_tree()
```

```python
mushrooms = pd.read_csv('agaricus-lepiota.csv')


target_attr = mushrooms.columns.tolist()[0]
attrs = mushrooms.columns.tolist()[1:]

id3 = ID3(criterion='gainratio')
id3.fit(mushrooms, target_attr, attrs)
id3.score(mushrooms, target_attr)
```

```python
id3.display_tree()
```
